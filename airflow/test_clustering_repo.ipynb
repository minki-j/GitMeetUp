```python
file_path = "/Users/minkijung/Downloads/no_filter_for_fork_and_forks.csv"
file_path_fork_false = "/Users/minkijung/Downloads/fork_false.csv"

# open with pandas
import pandas as pd
df = pd.read_csv(file_path_fork_false)
# print the number of rows
print(len(df))
df["packages_used"] = df["packages_used"].apply(
    lambda x: [pkg.strip() for pkg in x.split(",")]
)

df.head()
```

```python
# find rows that includes "minki-j" in html_url col
minki_j = df[df["html_url"].str.contains("minki-j")]
print(minki_j)
```

```python
packages = minki_j["packages_used"].to_list()
```

```python
packages
```

```python
my_packages = packages[1]
print(f"==>> my_packages: {my_packages}")

# Function to calculate the number of shared packages
def count_shared_packages(row, my_packages):
    return len(set(row) & set(my_packages))


# Apply the function to the DataFrame and create a new column with the counts
df["shared_packages_count"] = df["packages_used"].apply(
    count_shared_packages, my_packages=my_packages
)

# Sort the DataFrame by the count of shared packages in descending order and get the top 10 rows
top_rows = df.sort_values(by="shared_packages_count", ascending=False).head(10)

# print html and packages_used col only
top_rows[["html_url", "packages_used", "shared_packages_count"]]
```
